# Valuación de derivados

**Definición(Instrumento derivado)**

Un instrumento derivado es un contrato entre dos o más partes y cuyo valor se deriva a partir del valor de un activo subyacente.

**Definición(Opción sobre acciones)**

Una opción, representa el **derecho** de comprar (opción call) o vender (opción put) a un precio determinado (el precio strike) una unidad de la acción a la cual está referenciada.

**Definición(Opción europea)**

Una opción **europea** es una opción que se puede ejercer únicamente en una fecha determinada.

**Definición(Opción americana)**

Una opción **americana** es una opción que se puede ejercer en cualquier fecha dentro de un periodo de tiempo determinado.

**Definición(Payoff de opciones europeas)**

Sea $T$ el momento en que se estable que la opción se puede ejecutar y sea $K$ el precio strike. El payoff de una opción call europea con subyacente $S$, está dado por:

$$
Call_{T}(S,K) = \max(S_T - K, 0)
$$

Por otra parte, el payoff de una opción put europea es

$$
Put_{T}(S,K) = \max(K - S_T, 0)
$$

Vemo así que un call es una apuesta a una tendencia a la alza en los precios, mientras que un put es una apuesta a una tendencia a la baja.

## Modelo de Black-Scholes-Merton

**Definición(Movimiento browniano)**

Un movimiento browniano (formalmente un proceso de Wiener) es un proceso estocástico $\{W\}_{t\geq 0}$ que satisface lo siguiente:

1. W_0 = 0.

2. El mapeo $t \rightarrow W_t$ es una función continua.

3. Los incrementos $\{W_{t_1} - W_{t_0}, W_{t_2} - W_{t_1}, \ldots, W_{t_k} - W_{t_{k-1}} \}$ son variables aleatorias independientes.

4. Para $s < t$, $W_t - W_s \sim N(0, \sigma^2 = t - s)$.

En el modelo de Black-Scholes-Merton, se tiene un mercado en el cual existen únicamente tres instrumentos:

1. Acción $S$.

2. Cuenta de ahorro, $B_t$ que gana una tasa libre de riesgo $r$.

$$
\dfrac{B_t}{B_t} = rdt \implies B_t = e^{rt}
$$

3. Instrumento derivado $V$ cuyo subyacente es $S$.

**Definición(Movimiento browniano geométrico)**

Sea $W_t$ un movimiento browniano y sean $r, \sigma$ constantes con $\sigma > 0 $. Un movimiento browniano geométrico, $S$, es un proceso que satisface la siguiente ecuación diferencial estocástica

$$
dS_t = rS_tdt + \sigma S_tdW_t
$$

De manera equivalente

$$
\dfrac{dS_t}{S_t} = rdt + \sigma dW_t
$$

En el modelo de Black-Scholes, la acción se modela através de un proceso browniano geometríca, con el parámetro $r$ representando la tasa libre de riesgo (instantanea) y $\sigma$ representa la volatilidad (instantanea) de la acción.

**Teorema(Fórmula de Itô, funciones de un movimiento browniano)**

Sea $V(t,x)$ una función con derivadas parciales $V_t, V_x, V_{xx}$ continuas y sea $W_t$ un movimiento browniano, entonces para toda $T\geq0$ se tiene

$$
V(T,W_T) = V(0,W_0) + \int_{0}^{T}V_{t}(t,W_t)dt + \int_{0}^{T}V_{x}(t, W_t)dW_t + \dfrac{1}{2}\int_{0}^{T}V_{xx}(t,W_t)dt
$$

o en forma diferencial

$$
dV(T,W_T) = \dfrac{\partial V(t,W_t)}{\partial t}dt + \dfrac{\partial V(t,W_t)}{\partial x}dW_t + \dfrac{1}{2} \dfrac{\partial^{2}V(t,W_t)}{\partial x^2}dt
$$

**Teorema(Fórmula de Itô, funciones de un proceso de Itô)**

Si $S$ es un proceso que satisface la ecuación diferencial estocástica

$$
dS_t = rS_tdt + \sigma S_tdW_t
$$

y si $f$ es una función con derivadas parciales $f_t, f_x, f_xx$ continuas, entonces, en forma diferencial, la fórmula de Itô de $f(t,S_t)$ está dada por:

$$
df(t,S) = f_{t}(t,S_t)dt + f_{x}(t,S_t)dS_t + \dfrac{1}{2}f_{xx}(t,S_t)dS_t dS_t
$$

en donde el término $dS_t dS_t$ se obtiene utilizando la  siguiente tabla

| |$dW_t$|$dt$|
|-|----|--|
|$dW_t$|$dt$|$0$|
|$dt$|$0$|$0$|

Bajo los supuestos del modelo de Black-Scholes, el precio del derivado satisface la siguiente ecuación diferencial parcial

$$
\dfrac{\partial V}{\partial t} + \dfrac{1}{2} \sigma^2 S^2 \dfrac{\partial^2 V}{\partial S^2} + rS\dfrac{\partial V}{\partial S} - rV = 0
$$

**Fórmula analítica para valuar opciones call y put europeas bajo el modelo de Black-Scholes**

Bajo el modelo de Black-Scholes, el precio, en el tiempo $t$, de un call del tipo europeo está dado por

$$
Call(S,K,t,T,r,\sigma) = S_t\mathbf{N}(d_1) - e^{-r(T-t)}K\mathbf{N}(d_2)
$$
en donde

* $\mathbf{N}$ Es la función de distribución acumulada de una variable aleatoria $N(0,1)$.

* $d_1 = \dfrac{ \ln \dfrac{S_t}{K} +  (r + 0.5\sigma^2) (T-t) }{\sigma \sqrt{T-t}}$

* $d_2 = \dfrac{ \ln \dfrac{S_t}{K} +  (r - 0.5\sigma^2) (T-t) }{\sigma \sqrt{T-t}}$

* $T$ es la fecha de vencimiento (generalmente en años).

* $t$ es la fecha de valuación.

* $r$ es la tasa libre de riesgo

* $\sigma$ volatilidad del subyacente

* $K$ precio strike.

En el caso de un put, se tiene

$$
Put(S,K,t,T,r,\sigma) = e^{-r(T-t)}K\mathbf{N}(-d_2) - S_t\mathbf{N}(-d_1)
$$

**Teorema(Valuación neutral al riesgo)**

Bajo ciertas condiciones, el precio, en el tiempo $t$, de un producto financiero derivado con función de payoff $H$ y cuyo subyacente es $S$ está dado por

$$
Precio_H(S,t) = e^{-(T-t)}E_{Q}[H(S,T)]
$$

en donde $Q$ es una medida de probabilidad que cumple ciertas condiciones.

En particular para un call europeo

$$
Call(S,K,t,T,r,\sigma) = e^{-(T-t)}E_{Q}[\max(S_T - K, 0)]
$$

## Esquema de Euler para simular procesos estocásticos


Para simular el proceso dado por la ecuación diferencial estocástica

$$
dS_t = rS_tdt + \sigma S_t dW_t
$$

dentro de un intervalo $[t, T]$ se procede de la siguiente manera
1. Crear una partición, $\Pi = \{t_0 = t, t_1, \ldots, t_N = T \}$ del intervalo $[t, T]$ dividiéndolo en $N$ partes iguales y haciendo $\delta = \dfrac{T-t}{N}$

2. Sustituir $dt$ por $\delta$ y $dW_t$ por $\sqrt{\delta}Z$ en donde $Z \sim N(0,1)$.

3. Calcular los siguientes precios utilizado la discretización

$$
S_{t_{i+1}} = S_{t_{i}} + r S_{t_{i}} \delta + \sigma S_{t_{i}}\sqrt{\delta}Z_{i+1} 
$$
